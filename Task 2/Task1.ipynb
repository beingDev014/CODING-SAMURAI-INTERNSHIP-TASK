```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area, df.price, color='blue', marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python

reg.predict(pd.DataFrame([[3300]], columns=['area']))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y = m*x+b
135.78767123*3300+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='blue', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='red')
plt.show()
```

```python
d = pd.read_csv("areas.csv")
d.head(3)
```

```python
d['prices'] = p
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')

plt.scatter(df['year'], df['per capita income (US$)'], color='blue', marker='+')
plt.show()
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict(pd.DataFrame([[2020]], columns=['year']))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y = m*x+b
828.46507522*2020+(-1632210.7578554575)
```

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
pred_2020 = reg.predict([[2020]])[0]

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color='blue', marker='+')
plt.plot(df['year'], reg.predict(df[['year']]), color='red')

plt.scatter(2020, pred_2020, color='green', marker='o')  # Mark 2020 prediction
plt.annotate(f'2020: ${pred_2020:,.2f}', 
             xy=(2020, pred_2020), 
             xytext=(2010, pred_2020 + 2000),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```
